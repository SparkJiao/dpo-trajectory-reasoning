```python
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
import json
import torch

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

def plot_histogram(data, bins=10, xlabel="Value", ylabel="Frequency", title="Histogram"):
    plt.hist(data, bins=bins, edgecolor='black', alpha=0.7)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
data = np.random.randn(100)
plot_histogram(data, bins=10)
```

```python
def estimate_value(file):
    data = json.load(open(file, "r"))
    state_id2values = defaultdict(dict)
    for item in data:
        idx = item["id"]
        idx, state_id = idx.split("_")
        v = 0
        flag = True
        for res, p in zip(item["response"], item["pred"]):
            if res == "":
                flag = False
                break
            if p == "":
                continue
            if ord(p.strip()) - ord("A") == item["label"]:
                v += 1
        if not flag:
            continue
        state_id2values[int(idx)][int(state_id)] = (v, item["text"].split("Thought 1:")[1].strip())

    print(len(state_id2values))
    # max_res_num = max([len(x) for x in state_id2values.values()])
    # print(max_res_num)
    # state_id2values = {k: v for k, v in state_id2values.items() if len(v) == max_res_num}
    print(len(state_id2values))
    return state_id2values
```

```python
state_id2values = estimate_value("trajectories/turbo-sft-v2.0/logiqav2-train.full.qa.react.v1.0.0shot.inter.ver2.0.rs0.2.r0.6.split-0.sample3.json")
```

```python
data = json.load(open("trajectories/eval_predictions_rank3.json"))

def logit2prob(logits):
    probs = torch.softmax(logits, dim=-1)
    probs = probs[:, 2] + probs[:, 3]
    return probs

cnt = 0
rewards = []
for item in data:
    if "Finish[" in item["response"].split("\n")[-1]:
        cnt += 1
        # rewards.append(item["reward"])

        ending_logits = torch.tensor(item["ending_logits"])
        probs = logit2prob(ending_logits)
        acc_probs = 1.0
        res = []
        for prob in probs:
            acc_probs = acc_probs * prob
            res.append(acc_probs.item())

        rewards.extend(res)
        

plot_histogram(rewards, bins=10)
print(cnt)
```

```python
def process_response(response: str):
    lines = response.split("\n")
    lines = list(filter(lambda x: x[1].startswith("Thought ") or x[1].startswith("Action ") or x[1].startswith("Observation "), enumerate(lines)))
    return lines

def test(response: str, max_seq_length: int):
    steps = process_response(response)
    raw_lines = response.split("\n")
    endings = []
    resp_start = False
    for step_id, step in steps:
        if not resp_start:
            if step.startswith("Thought 1:"):
                resp_start = True
            else:
                continue
        # partial_traj = "\n".join(raw_lines[:(step_id + 1)])
        endings.append(0)

    assert resp_start, response
    assert len(endings) > 0, (response, steps)
    assert len(endings) == len(process_response("Thought 1:" + response.split("Thought 1:")[1])), (response, steps)
    return endings

data = json.load(open("trajectories/eval_predictions_rank3.json"))
cnt_a = 0
cnt_b = 0
samples_a = []
samples_b = []
for item in data:
    resp = "Thought 1: " + item["response"]
    steps = process_response(resp)
    if len(steps) != len(item["ending_logits"]):
        cnt_a += 1
        samples_a.append((len(steps), len(item["ending_logits"])))
    steps = test(item["input"], 4096)
    if len(steps) != len(item["ending_logits"]):
        cnt_b += 1
        samples_b.append((len(steps), len(item["ending_logits"])))
```
